```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv('stock.csv', sep=';')

# Mengatur ukuran grafik
plt.figure(figsize=(10, 6))

# Mengatur gaya Seaborn
sns.set(style="whitegrid")

# Menampilkan distribusi atribut Purchase Price
sns.histplot(data=df, x="Purchase Price", kde=True, color='blue')
plt.title('Distribusi Purchase Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Menampilkan distribusi atribut Current Price
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Current Price", kde=True, color='green')
plt.title('Distribusi Current Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Menampilkan distribusi atribut Gain/Loss
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Gain/Loss", kde=True, color='red')
plt.title('Distribusi Gain/Loss')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv('stock.csv', sep=';')

# Mengurutkan data berdasarkan Dividend/Share secara menurun
sorted_df = df.sort_values('Dividend/share', ascending=False).head(5)

# Menggabungkan kolom Stock Name dan Symbol menjadi satu kolom baru
sorted_df['Stock'] = sorted_df['Stock Name'] + ' (' + sorted_df['Symbol'] + ')'

# Menghapus karakter non-numerik dan mengonversi kolom 'Dividend/share' menjadi tipe data numerik
sorted_df['Dividend/share'] = sorted_df['Dividend/share'].str.replace(r'\$', '', regex=True).astype(float)

# Membuat bar plot menggunakan Seaborn dengan warna merah pada batang pertama
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='Stock', y='Dividend/share', palette=['red', 'blue', 'darkblue', 'aqua', 'lightblue'] + sns.color_palette()[1:])
plt.title('Top 5 Stocks by Dividend/Share')
plt.xlabel('Stock')
plt.ylabel('Dividend/Share ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv('stock.csv', sep=';')

# Mengurutkan data berdasarkan Market Value secara menurun
sorted_df = df.sort_values('Market Value', ascending=False).head(5)

# Menggabungkan kolom Stock Name dan Symbol menjadi satu kolom baru
sorted_df['Stock'] = sorted_df['Stock Name'] + ' (' + sorted_df['Symbol'] + ')'

# Menghapus karakter non-numerik (",", "$") dan mengonversi kolom 'Market Value' menjadi tipe data numerik
sorted_df['Market Value'] = sorted_df['Market Value'].str.replace('[$,]', '', regex=True).astype(float)

# Membuat bar plot menggunakan Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='Stock', y='Market Value', palette='viridis')
plt.title('Stocks by Market Value')
plt.xlabel('Stock')
plt.ylabel('Market Value ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Membaca file CSV
df = pd.read_csv('stock.csv', sep=';')

# Mengurutkan data berdasarkan Annual Yield secara menurun
sorted_df = df.sort_values('Annual Yield', ascending=False).head(7)

# Menggabungkan kolom Stock Name dan Symbol menjadi satu kolom baru
sorted_df['Stock'] = sorted_df['Stock Name'] + ' (' + sorted_df['Symbol'] + ')'

# Mengubah kolom Annual Yield menjadi tipe data numerik
sorted_df['Annual Yield'] = sorted_df['Annual Yield'].str.replace('%', '').astype(float)

# Membuat bar plot menggunakan Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=sorted_df, x='Stock', y='Annual Yield', palette='viridis')
plt.title('Top 7 Stocks by Annual Yield')
plt.xlabel('Stock')
plt.ylabel('Annual Yield (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
